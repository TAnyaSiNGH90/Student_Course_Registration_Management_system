This dataset contains records of **1,545 college students**, capturing their demographics, academic performance, learning engagement, and risk levels. It includes attributes such as **age, gender, major, GPA, attendance, course load, assignment submission, LMS logins, video completion, and enrollment status**. The dataset is designed for analyzing student performance patterns, identifying at-risk learners, and supporting data-driven decision-making in academic management.

```python
plt.figure(figsize=(6,4))
plt.hist(std['GPA'], bins=20, color='skyblue', edgecolor='black')
plt.title("GPA Distribution")
plt.xlabel("GPA")
plt.ylabel("Number of Students")
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(6,4))
plt.scatter(std['attendance_rate'], std['GPA'], alpha=0.5, color='blue', edgecolors='black')
plt.title("Attendance Rate vs GPA")
plt.xlabel("Attendance Rate")
plt.ylabel("GPA")
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(6,4))
plt.scatter(std['assignment_submission_rate'], std['GPA'], alpha=0.5, color='green', edgecolors='black')
plt.title("Assignment Submission Rate vs GPA")
plt.xlabel("Assignment Submission Rate")
plt.ylabel("GPA")
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10,7))
corr = std.corr(numeric_only=True)
sns.heatmap(corr, annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Correlation Heatmap of Student Data")
plt.tight_layout()
plt.show()
```

* Average attendance rate: ~79%.

* Average assignment submission rate: ~75%.

* Average video completion rate: ~75%.







Engagement & Performance ---

 Key Takeaways:

* Arts has the largest enrollment, but all majors are fairly balanced.
